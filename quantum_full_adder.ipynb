```python
q = QuantumRegister(8)  # Increase the number of qubits to accommodate the third input
c = ClassicalRegister(2)
qc = QuantumCircuit(q, c)
```

```python
def full_adder(a, b, cin, d, e, s, g, cout):
    qc.ccx(q[a], q[b], q[d])
    qc.cx(q[a], q[e])
    qc.cx(q[b], q[e])
    qc.cx(q[cin],q[s])
    qc.cx(q[e], q[s])
    qc.ccx(q[cin], q[e], q[g])
    qc.ccx(q[d], q[g], q[cout])


# Inputs: a, b, cin
# Outputs: cout, s
```

```python
full_adder(0, 1, 2, 3, 4, 5, 6, 7)

# Measurements
qc.measure(q[5], c[0])  # Sum
qc.measure(q[7], c[0])  # Cout
```

```python
qc.draw(output='mpl') # draws the circuit
```
